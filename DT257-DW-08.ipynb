## Cleaning Dataset

```python
#dropping unnecessary columns

data= review_data.drop(['link','Unnamed: 0' ], 1)
```

```python
#checking that they are dropped

data
```

```python
#dropping any rows that have the none value in location

data.drop(data[data['location']=='none'].index, inplace= True)
```

In these next few we are replacing any nones in the stars features with Nan

```python
# replacing any nones in the stars feature with Nan

data['work-balance-stars'].replace('none', np.NaN, inplace=True)
```

```python
data['culture-values-stars'].replace('none', np.NaN, inplace=True)
```

```python
data['carrer-opportunities-stars'].replace('none', np.NaN, inplace=True)
```

```python
data['comp-benefit-stars'].replace('none', np.NaN, inplace=True)
```

```python
data['senior-mangemnet-stars'].replace('none', np.NaN, inplace=True)
```

```python
#checking to see if these features none values are being read as null

data.isnull().sum()
```

```python
# checking the data types, as this is the next roadblock to statistical analysis

data.astype({'overall-ratings': 'int32'}).dtypes
```

In these next cells we fill all the Nan values with 0's and then change the datatype of each feature so that statistics can be done on the feature

```python
data['work-balance-stars'].fillna(0)
data['work-balance-stars']=  data['work-balance-stars'].astype(float, errors = 'raise')
```

```python
data['culture-values-stars']= data['culture-values-stars'].fillna(0)
```

```python
data['culture-values-stars']= data['culture-values-stars'].astype(float, errors = 'raise')
```

```python
data['carrer-opportunities-stars']= data['carrer-opportunities-stars'].fillna(0)
data['carrer-opportunities-stars']= data['carrer-opportunities-stars'].astype(float, errors = 'raise')
```

```python
data['comp-benefit-stars']= data['comp-benefit-stars'].fillna(0)
data['comp-benefit-stars']= data['comp-benefit-stars'].astype(float, errors = 'raise')
```

```python
data['senior-mangemnet-stars']= data['senior-mangemnet-stars'].fillna(0)
data['senior-mangemnet-stars']= data['senior-mangemnet-stars'].astype(float, errors = 'raise')
```

```python
# separating those from Dublin out to their own Dataframe

Dublin = df[df['Location']=='Dublin, Co. Dublin (Ireland)']
Dublin
```

```python
# making a new dataframe with the dublin companies and their means

Dub_Agg= Dublin.groupby('Company').aggregate(['mean']).sort_values('Company')
Dub_Agg
```

```python
# To avoid the mutliindex error, we rename the columns, reset the index, make a new Dataframe, drop helpful mean

Dub_Agg.columns= ['overall_mean', 'work_life_mean', 'culture_mean', 'career_mean', 'benefits_mean', 'management_mean', 'helpful_mean']
Dub_Agg = Dub_Agg.reset_index()
Dub= pd.DataFrame(Dub_Agg)
Dub= Dub.drop(['helpful_mean'], 1)
Dub
```

```python
# sorting the values by their means, so that the highest scoring companies are at the top

Dub_final= Dub.sort_values(['overall_mean', 'work_life_mean', 'culture_mean', 'career_mean', 'benefits_mean', 'management_mean'], ascending=False)
Dub_final
```

```python
# separating those from Cork out to their own Dataframe

Cork= df[df['Location']=='Cork (Ireland)']
Cork
```

```python
# making a dataframe for just the mean of the various star/rating features & dropping found helpful

Cork_Agg= Cork.groupby('Company').aggregate(['mean']).sort_values('Company')
cork_final= Cork_Agg.drop(['Found Helpful'], 1)
cork_final
```

```python
# changing the name of the columns, and sorting the values for the overall cork company statistics

cork_final.columns= ['overall_mean', 'work_life_mean', 'culture_mean', 'career_mean', 'benefits_mean', 'management_mean']
CRK= cork_final.sort_values(['overall_mean', 'work_life_mean', 'culture_mean', 'career_mean', 'benefits_mean', 'management_mean'], ascending=False)
CRK
```

```python
# importing a dataset with the information about various ICT companies in Ireland

IRL_ICT= pd.read_csv('IRL_ICT.csv')
IRL_ICT
```

```python
# web scraping from a html file, in this case the CSO, so we know the commuting times 

commute_IRL= pd.read_html("https://www.cso.ie/en/releasesandpublications/ep/p-cp6ci/p6cii/p6td/")
commute_IRL= commute_IRL[1]
commute_IRL
```

```python
# web scraping the cost of living for Dublin

cost_of_living_dulbin= pd.read_html("https://www.expatarrivals.com/europe/ireland/cost-living-ireland")
cost_of_living_dulbin= cost_of_living_dulbin[0]
cost_of_living_dulbin
```

```python
# Doing the same for Cork

cost_of_living_cork= pd.read_html("https://www.travelsafe-abroad.com/cost-of-living/ireland/cork/")
cost_of_living_cork= cost_of_living_cork[0]
cost_of_living_cork
```

```python
# sliming the Cork entries that are relevant to us and provide the best information that is needed

cost_cork= cost_of_living_cork.loc[[9, 18, 19, 20], :]
cost_cork= cost_cork.transpose()
cost_cork.columns=['1 Bed centre', '1 Bed outside', '3 bed centre', 'Utilities']
cost_cork= cost_cork.drop(0)
cost_cork
```

```python
# Doing the same with the Dublin information

cost_dublin= cost_of_living_dulbin.loc[[1, 2, 3, 4, 21], :]
cost_dublin= cost_dublin.transpose()
cost_dublin.columns=['1 Bed centre', '3 bed centre', '1 Bed outside', '3 bed outside', 'Utilities']
cost_dublin= cost_dublin.drop(0)
cost_dublin
```

```python
# boiling down the commuting times to the two cities we want 

commute_IRE= commute_IRL.loc[[8, 27]]
commute_IRE = commute_IRE.reset_index(drop=True)
commute_IRE
```

```python
# separating out the Cork commuting time

Cork_commute= commute_IRE.loc[[1]]
Cork_commute
```

```python
# same with the Dublin commute

Dub_commute= commute_IRE.loc[[0]]
Dub_commute
```

```python
# now putting the commute and the cost of living together

Dub= pd.concat([Dub_commute, cost_dublin])
Dub
```

```python
# issues with the concat meant that some values had to changed manually

Dub.iloc[1, 0]= 'Dublin City'
Dub.iloc[1, 1]= '28.9'
Dub= Dub.drop(0)
Dub
```

```python
# now concating the Cork commute and cost of living

Cork= pd.concat([Cork_commute, cost_cork])
Cork
```

```python
# again concat causing issues so cleaning this up

Cork.iloc[1, 0]= 'Cork City'
Cork.iloc[1, 1]= '22.1'
Cork= pd.DataFrame(Cork.iloc[1])
Cork= Cork.transpose()
Cork
```

```python
#Now combining all these elements together to make a new dataset for cost of living and commute for both Dublin & Cork

Ireland= pd.concat([Cork, Dub])
Ireland
```

```python
# Since the 3 bed outside is only a feature in Dublin city we decided to drop it as no comparison would be possible

Ireland= Ireland.drop(labels='3 bed outside', axis=1)
Ireland
```

```python
# resetting the index as both records were coming up as one

Ireland= Ireland.reset_index(drop=True)
Ireland
```

```python
# Since only really Apple is in Cork for software engineering, I decided to narrow these down to two
# would have done it to just Apple but this really messed up the indexing in the next stages

Cork_comp= CRK.iloc[1:3]
Cork_comp= pd.DataFrame(Cork_comp)
Cork_comp
```

```python
# dropping Apple from the Dublin dataframe, since the factory is in Cork and all that is in Dublin I believe is Apple shops

Dub_comp = Dub_final.drop(1)
Dub_comp
```

```python
# merging the the Cork and Dublin Companies together into a new dataset

Ireland_comp= pd.merge(Dub_comp, Cork_comp, on='Company', how='outer')
Ireland_comp
```

```python
# Now moving the Apple features over by hand to their appropriate cells

Ireland_comp['overall_mean_x'][4]= 3.502326
Ireland_comp['work_life_mean_x'][4]= 3.069048
Ireland_comp['culture_mean_x'][4]= 3.274419
Ireland_comp['career_mean_x'][4]= 3.013953
Ireland_comp['benefits_mean_x'][4]= 3.758140
Ireland_comp['management_mean_x'][4]= 2.988372
Ireland_comp
```
